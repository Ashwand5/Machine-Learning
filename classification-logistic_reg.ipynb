# Logistic Regression

# Data Pre-processing Step
# importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# importing dataset

```python
data_set = pd.read_csv('user_data.csv')
```

# Extracting Independent and dependent Variables

```python
x = data_set.iloc[:, [2, 3]].values
y = data_set.iloc[:, 4].values
```

# Splitting the dataset into training and test set.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
```

# Fitting Logistic Regression to the training set

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(x_train, y_train)
```

# Predicting the test set result

```python
y_pred = classifier.predict(x_test)
```

# Creating the Confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

# Visualizing the training set result

```python
# Visualizing the training set result
from matplotlib.colors import ListedColormap

x_set, y_set = x_train, y_train
x1, x2 = np.meshgrid(np.arange(start=x_set[:, 0].min() - 1, stop=x_set[:, 0].max() + 1, step=0.01),
                     np.arange(start=x_set[:, 1].min() - 1, stop=x_set[:, 1].max() + 1, step=0.01))

# Choose a more standard color map
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA'])

plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha=0.75, cmap=cmap_light)

plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                edgecolor='k', marker='o', s=50, label=j)  # Use black edgecolor for better visibility

plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

# Visulaizing the test set result

```python
# Visulaizing the test set result
from matplotlib.colors import ListedColormap
x_set, y_set = x_test, y_test
x1, x2 = np.meshgrid(np.arange(start=x_set[:, 0].min() - 1, stop=x_set[:, 0].max() + 1, step=0.01),
                     np.arange(start=x_set[:, 1].min() - 1, stop=x_set[:, 1].max() + 1, step=0.01))

# Choose different colors for the contour plot
cmap_contour = ListedColormap(['lightcoral', 'lightgreen'])

plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha=0.75, cmap=cmap_contour)
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

# Choose different colors for scatter points
cmap_scatter = ListedColormap(['red', 'green'])
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c=cmap_scatter(i), label=j)

plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score, accuracy_score

# R-squared (R2) score
r2 = r2_score(y_test, y_pred)
print(f'R-squared (R2) score: {r2}')

# Accuracy score
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy score: {accuracy}')
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Precision
precision = precision_score(y_test, y_pred)
print(f'Precision: {precision}')

# Recall
recall = recall_score(y_test, y_pred)
print(f'Recall: {recall}')

# F1-Score
f1 = f1_score(y_test, y_pred)
print(f'F1-Score: {f1}')
```
